```python
total = hospital_df['Department'].value_counts().sum()
percentage_values = (hospital_df['Department'].value_counts() / total) * 100

# Extract data for plotting
labels = hospital_df['Department'].unique()
sizes = hospital_df['Department'].value_counts()

# Create a pie chart
plt.figure(figsize=(10, 10))
plt.pie(sizes, labels=labels, autopct=lambda pct: '{:.1f}%'.format(pct), startangle=140)
plt.axis('equal')
plt.title('Pie Chart of Department')
plt.show()
```

```python
total = hospital_df['Type of Admission'].value_counts().sum()
percentage_values = (hospital_df['Type of Admission'].value_counts() / total) * 100

# Extract data for plotting
labels = hospital_df['Type of Admission'].unique()
sizes = hospital_df['Type of Admission'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct=lambda pct: '{:.1f}%'.format(pct), startangle=140)
plt.axis('equal')
plt.title('Pie Chart of Type of Admission')
plt.show()
```

```python
excluded_avaialble_extra_room = [11,0,12,13,14,24]  # List of extra rooms to exclude
filtered_df = hospital_df[~hospital_df['Available Extra Rooms in Hospital'].isin(excluded_avaialble_extra_room)]

# Plot the histogram using Seaborn

sns.histplot(data=filtered_df, x='Available Extra Rooms in Hospital', bins=25, color='blue')
plt.xlabel('Available Extra Rooms in Hospital')
plt.ylabel('Frequency')
plt.title('Histogram of Available Extra Rooms in Hospital (Excluding Outliers)')
plt.xlim(right=8)
plt.show()
```
